```python
import matplotlib.pyplot as plt

USA_DATA= pd.read_csv(r"us-marriages-divorces-1867-2014.csv",sep=",")
USA_DATA
```

```python
x_axis = USA_DATA[["Year"]]
figure, axis = plt.subplots()
axis.plot(x_axis,USA_DATA[["Marriages_per_1000"]])
axis.plot(x_axis,USA_DATA[["Divorces_per_1000"]])
axis.set_title('Marriages and divorces per capita in the U.S. between 1867 and 2014')
axis.set_xlabel('Year')
axis.set_ylabel('Marriages and Divorces per capita')
axis.legend(USA_DATA[["Marriages_per_1000","Divorces_per_1000"]])
```

# Question 7

```python
data=USA_DATA.loc[(USA_DATA['Year']==1900) | (USA_DATA['Year']==1950) | (USA_DATA['Year']==2000)]
barchart = data.groupby("Year").sum().plot.bar(figsize=(10,8))
barchart.set_title('number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.')
barchart.set_xlabel('Year')
barchart.set_ylabel('Marriages / Divorces')
barchart.legend()
```
